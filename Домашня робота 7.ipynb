**Зауваження.**
За бажання чи необхідності сайт можна змінити.
Тобто головна задача завдання - збір даних з певного сайту. А якого саме - не так важливо.

````python
import requests
from bs4 import BeautifulSoup
lis = {}
sl = []

url = 'https://privatbank.ua/'
link = requests.get(url)
soup = BeautifulSoup(link.content, 'lxml')
dop_products = soup.find_all('div', class_='column-menu g-c1') 
for dop_product in dop_products:
    c_list = dop_product.find_all('a')
    name = c_list[0].text
   # print(f'Назва: {name} >> {link}')
    link = 'https://privatbank.ua'+ c_list[0].get('href').replace('https://privatbank.ua','')
    href = c_list[0].get('href').replace('https://privatbank.ua','').replace('/','_')
    r = requests.get(link)
    soup_dop = BeautifulSoup(r.text, 'lxml')
    c1_list = soup_dop.find_all('div', class_="wr_inner")  
    for c_product in c1_list:
        sl = c_product.text.split()
    lis[href] = {'name': name,
                  'link': link,
                    'sl': sl,
                 'href': href}
    
from pprint import pprint
pprint(lis)
````
